```python
def generator_even_number(limit):
    for i in range(0, limit, 2):
        yield i
        
```

```python
def generator_powers_of_two():
     num = 0
     while True:
         num = num + 1
         yield 2 ** num
two = generator_powers_of_two()
two
```

```python
next(two)
```

```python
next(two)
```

```python
count = 0

for p in two:
    print(p)
    count = count + 1
    if count > 10:
        break
```
