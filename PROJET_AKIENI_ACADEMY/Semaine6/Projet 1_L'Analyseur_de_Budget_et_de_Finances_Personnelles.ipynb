# Projet 1 : L'Analyseur de Budget et de Finances Personnelles


## **L'objectif** : Créer un programme interactif dans le terminal permettant à un utilisateur d'enregistrer, de classer et d'analyser ses revenus et ses dépenses quotidiennes.

## Cahier des charges et Fonctionnalités
- Menu principal : Le programme tourne en boucle jusqu'à ce que l'utilisateur choisisse de quitter. Il propose d'ajouter une transaction, d'afficher le solde, ou de     voir un résumé par catégorie.

- Ajout d'une transaction : L'utilisateur saisit un montant, un type (revenu ou dépense) et une catégorie (ex: "Alimentation", "Loisirs", "Salaire").

- Sécurité des saisies : Le programme ne doit jamais planter. Si l'utilisateur tape "vingt" au lieu de "20", le programme affiche un message d'erreur clair et redemande la saisie.

- Analyse : Une option du menu permet d'afficher la dépense la plus importante et de calculer le pourcentage que représente chaque catégorie dans les dépenses totales.

```python
CATEGORIES_AUTORISEES = ["Alimentation", "Loisirs", "Transport", "Logement", "santé", "Facture", "Investissement", "Epargne"]
```

```python
def ajouter_transaction(transactions, montant, categorie, type_tx, categories_autorisees):
    """
    Vérifier si la catégorie est valide. Si oui, ajouter un dictionnaire {"montant": montant, "categorie": categorie, "type": type_tx} à la liste transactions.
        arg:
        - transactions (liste) : La liste de toutes les transactions existantes.
        - montant (float) : Le montant de la transaction (ex: 24.50)
        - categorie(str) : La catégorie saisie par l'utilisateur.
        - type_tx (str) : Soit "revenu", soit "depense".
        - categories_autorisees (tuple) : Le tuple contenant les catégories valides.
        return:
            True si la transaction a été ajoutée avec succès, False sinon. 
    """ 
    #Vérifier si la categorie n'est pas dans le tuple categories_autorisees.
        #Si c'est le cas, afficher un message d'erreur et renvoyer False.
    if categorie not in categories_autorisees:
        return False
        
    #Ajouter ce nouveau dictionnaire à la liste "transactions" (.append()) 
    else:
        nouvelle_transaction = {
            "montant": float(montant),
            "categorie": categorie,
            "type": type_tx
        }
        transactions.append(nouvelle_transaction)
        return True
    pass
```

```python
def calculer_solde(transactions):
    """Parcourir la liste et faire la somme des revenus moins la somme des dépenses.

    Args:
        transactions (_liste_): La liste des transactions.
    return:
        Retourne : float (le solde net).
    """
    
    solde = 0.0
    
    #Parcoure la liste des transactions
    for transaction in transactions:
        # Pour chaque transaction, vérifier son type
        # Si le type est "revenu", ajouter le montant au solde
        if transaction["type"] == "revenu":
            solde += transaction["type"]
            
        # Si le type est "depense", soustraire le montant du solde.
        elif transaction["type"] == "depense":
            solde -= transaction["type"]
    return solde
```

```python
def calculer_pourcentages_depenses(transactions):
    """
        Calculer le total des dépenses, puis déterminer la part (en %) de chaque catégorie de dépense.
    Args:
        transactions (liste) : La liste des transactions.
    Return: 
        Un dictionnaire {categorie: pourcentage} (ex: {"Alimentation": 45.0, "Loisirs": 55.0}).
    """
    totaux_par_categorie = {}
    total_depenses_global = 0.0
    
    # Parcourir toutes les transaction
    for transaction in transactions:
        # Attention : ignorez les transactions de type "revenu" !
        if transaction["type"] == "revenu":
            continue
        elif transaction["type"] == "depense":
            categorie = transaction["categorie"]
            montant = transaction["montant"]
    #calculer d'une part le total_depenses_global
            total_depenses_global += montant
            
            # 2. On cumule par catégorie dans le dictionnaire
            if categorie in totaux_par_categorie:
                totaux_par_categorie[categorie] += montant
            else:
                totaux_par_categorie[categorie] = montant
                
    #Sécurité division par zéro ---           
    if total_depenses_global == 0.0:
        return {}
    
    pourcentages = {}
    #Parcourir le dictionnaire "totaux_par_categorie". Pour chaque catégorie
    for categorie in totaux_par_categorie:
        # On récupère le montant cumulé de cette catégorie
        montant_categorie = totaux_par_categorie[categorie]
        
        # On calcule sa part en % et on l'arrondit à 1 chiffre après la virgule
        pourcentage_calcule = round((montant_categorie / total_depenses_global) * 100, 1)
        
        # On stocke le résultat dans le dictionnaire final
        pourcentages[categorie] = pourcentage_calcule
    return pourcentages
```

```python
def menu_principal():
    """
    Fonction principale qui gère l'interface utilisateur dans le terminal.
    Elle intègre la gestion des erreurs de saisie pour éviter tout plantage.
    """
    # On initialise une liste vide pour stocker l'historique des transactions
    transactions = []
    
    print("="*67)
    print(" BIENVENUE SUR VOTRE ANALYSEUR DE BUDGET ")
    print("="*67)
    
    while True:
        # Affichage du menu
        print("========================== MENU =============================")
        print("1. Enregistrer un REVENU ")
        print("2. Enregistrer une DÉPENSE ")
        print("3. Consulter le SOLDE global ")
        print("4. Analyser la RÉPARTITION des dépenses (%) ")
        print("5. Quitter le programme ")
        
        choix = input("\nChoisissez une option (1 à 5) : ").strip()
        
        # --- OPTION 1 & 2 : AJOUT D'UNE TRANSACTION ---
        if choix in ("1", "2"):
            type_tx = "revenu" if choix == "1" else "depense"
            while True:
                # Gestion de l'exception pour la saisie du montant
                try:
                    montant_saisi = input("Entrez le montant : ").strip()
                    montant = float(montant_saisi)

                    if montant <= 0:
                        print("Le montant doit être strictement supérieur à 0.")
                        # Retour au début de la boucle
                    else:
                        # Le montant est valide ! On sort de la boucle du montant
                        break
                except ValueError:
                    print(f"Erreur : '{montant_saisi}' n'est pas un nombre valide. Saisie annulée.")
                    
            while True:
                # Affichage des catégories autorisées pour guider l'utilisateur
                print(f"\nCatégories valides : {', '.join(CATEGORIES_AUTORISEES)}")
                categorie = input("Entrez la catégorie correspondante : ").strip()

                # Appel de la fonction de validation et d'ajout
                succes = ajouter_transaction(transactions, montant, categorie, type_tx, CATEGORIES_AUTORISEES)

                if succes:
                    print(f"Transaction enregistrée : {type_tx.upper()} de {montant:.2f}FCFA dans '{categorie}'.")
                    break
                else:
                    print("Échec de l'enregistrement. (Vérifiez l'orthographe de la catégorie).")
                
        # --- OPTION 3 : AFFICHAGE DU SOLDE ---
        elif choix == "3":
            solde = calculer_solde(transactions)
            print("="*67)
            print(f"   SOLDE ACTUEL : {solde:.2f} FCFA")
            print("="*67)
            if solde < 0:
                print("Alerte rouge ! Vous dépensez plus que ce que vous gagnez.")
            elif solde > 0:
                print("Félicitations, vos comptes sont dans le vert !")
            else:
                print("Parfaitement à l'équilibre (0.00 €).")
                
        # --- OPTION 4 : POURCENTAGES DES DÉPENSES ---
        elif choix == "4":
            statistiques = calculer_pourcentages_depenses(transactions)
            
            print("======================= STATISTIQUES DES DÉPENSES ================================")
            if not statistiques:
                print("Aucune dépense enregistrée pour le moment. Rien à analyser !")
            else:
                for cat, pourcentage in statistiques.items():
                    print(f"• {cat} : {pourcentage}%")
                    
        # --- OPTION 5 : QUITTER ---
        elif choix == "5":
            print("\nMerci d'avoir utilisé l'Analyseur de Budget. À bientôt et bonne gestion !")
            break
            
        # --- CAS DE SAISIE INCORRECTE DU MENU ---
        else:
            print("Option invalide. Veuillez entrer un chiffre entre 1 et 5.")
```

```python
menu_principal()
```
